# Visualization

```python
 # 1. Histogram
 import matplotlib.pyplot as plt

# Histogram of one feature (example: sepal length)
plt.hist(data.data[:, 0], bins=10, edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
# 2. Scatter Plot created using Matplotlib instead of Seaborn

import matplotlib.pyplot as plt

# Plot each species with its own label
colors = ['red', 'green', 'blue']
species_names = data.target_names

for i, species in enumerate(species_names):
    species_data = data.data[data.target == i]
    plt.scatter(species_data[:, 0], species_data[:, 1], 
                c=colors[i], label=species, alpha=0.7)

plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.legend()  # This adds the legend
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
# 3. correlation heatmap
import seaborn as sns

df = pd.DataFrame(data.data, columns=data.feature_names)
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Feature Correlations')
plt.show()
```

```python
# 4. Scatter Plot created using seaborn
import pandas as pd
import seaborn as sns

df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = data.target_names[data.target]

sns.scatterplot(data=df, x=data.feature_names[0], y=data.feature_names[1], 
                hue='species')
plt.show()  # Legend appears automatically
```
